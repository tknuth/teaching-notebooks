```python
import altair as alt
import pandas as pd

from timeit import timeit

def sum_naive(n):
    value = 0

    for i in range(1, n+1):
        value += i
    
    return value

def sum_gauss(n):
    return n*(n+1)/2

funcs = {
    "sum_naive": sum_naive,
    "sum_gauss": sum_gauss
}

df = pd.DataFrame({
    k: [timeit(lambda: v(n), number=10000) for n in range(50)] 
    for k,v in funcs.items()
})

df = df.reset_index().melt(id_vars="index", value_vars=funcs.keys())

alt.Chart(df.reset_index()).mark_line().encode(
    x="index",
    y="value",
    color="variable"
)
```
