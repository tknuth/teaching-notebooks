```python
from scipy.special import factorial
import pandas as pd
import numpy as np
import altair as alt

alt.data_transformers.disable_max_rows()

funcs = {
    "lg(n)": np.log2,
    "sqrt(n)": np.sqrt,
    "n": lambda n: n,
    "n*lg(n)": lambda n: n * np.log2(n),
    "n^2": lambda n: n ** 2,
    "2^n": lambda n: 2 ** n,
    "n!": factorial,
}

x = np.arange(1, 10, 0.01)

df = pd.DataFrame({"n": x, **{k: v(x) for k, v in funcs.items()}})
df = df.applymap(lambda x: np.nan if x > 20 else x)
df = df.melt(id_vars="n")
df = df.rename({"variable": "function", "value": "duration"}, axis=1)

alt.Chart(df).mark_line().encode(
    x="n",
    y="duration",
    color="function"
)
```
