```python
import pandas as pd
import numpy as np
import itertools as it
from scipy.special import factorial
from scipy.optimize import newton

pd.set_option('display.float_format', '{:.2E}'.format)

funcs = {
    "lg(n)": np.log2,
    "sqrt(n)": np.sqrt,
    "n": lambda n: n,
    "n*lg(n)": lambda n: n*np.log2(n),
    "n^2": lambda n: n**2,
    "n^3": lambda n: n**3,
    "2^n": lambda n: 2**n,
    "3^n": lambda n: 3**n,
    "n!": factorial
}

ms = (pd.Series({
    "1 Sekunde": 1,
    "1 Minute": 60,
    "1 Stunde": 60**2,
    "1 Tag": 60**2*24,
    "1 Monat": 60**2*24*30,
    "1 Jahr": 60**2*24*365,
    "1 Jahrhundert": 60**2*24*365*100
})*10**6).astype("float64")

df = pd.DataFrame(index=funcs.keys(), columns=ms.index)

df.loc["n", :] = ms
df.loc["sqrt(n)"] = ms**2
df.loc["n^2"] = np.sqrt(ms)
df.loc["n^3"] = np.power(ms, 1/3)
df.loc["2^n"] = np.log(ms)/np.log(2)
df.loc["3^n"] = np.log(ms)/np.log(3)
df.loc["lg(n)"] = 2**ms

def factorial_limit(y):
    for n in it.count():
        if factorial(n) > y:
            return n-1

df.loc["n!"] = ms.apply(factorial_limit)
df.loc["n*lg(n)"] = ms.apply(lambda y: newton(lambda n: n*np.log2(n)-y, y))

df
```
