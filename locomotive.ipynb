```
import numpy as np
import matplotlib.pyplot as plt

hypos = np.arange(1,1000)

# plt.rcParams["figure.figsize"] = (10,5)

def likelihood(data, hypo):
    if hypo < data:
        return 0
    return 1/hypo

lhs = np.array([likelihood(60, x) for x in hypos])
print((lhs/lhs.sum()*hypos).sum())
plt.plot(hypos, lhs)
```

```
class Locomotive():
  def __init__(self, n):
    self.probas = [1/n for i in range(n)]

  def likelihood(self, data, hypo):
    if hypo < data:
        return 0
    return 1/hypo

  def normalize(self):
    s = sum(self.probas)
    for i, p in enumerate(self.probas):
      self.probas[i] /= s

  def update(self, data):
    for d in data:
      for hypo, p in enumerate(self.probas):
        self.probas[hypo] = self.likelihood(d, hypo)*p
    self.normalize()
    return self

plt.plot(Locomotive(1000).update([30,60,90]*3).probas)
```
