```python
# Bei welchen Flughäfen fehlen Angaben zum Staat?
airports[airports.state.isna()]
```

```python
airports[airports.name.str.contains("International")]
```

```python
# Doppelte Namen
airports[airports.duplicated(subset=["name"], keep=False)].sort_values("name")
```

```python
# Flughäfen, die in New York liegen oder mit "New " beginnen
# Wofür werden die Klammern benötigt?
# | ist "oder"
# & ist "und"
airports[airports.name.str.startswith("New ") | (airports.state == "NY")]
```
