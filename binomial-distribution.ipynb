```python
import numpy as np
import pandas as pd
from scipy.stats import binom
```

```python
x = range(10)
b1 = pd.DataFrame({
    "x": x,
    "params": "B(10, 0.5)",
    "PMF": [binom.pmf(k,10,0.5) for k in x],
    "CDF": [binom.cdf(k,10,0.5) for k in x]
    
})
b2 = pd.DataFrame({
    "x": x,
    "params": "B(1000, 0.5^10)",
    "PMF": [binom.pmf(k,1000,0.5**10) for k in x],
    "CDF": [binom.cdf(k,1000,0.5**10) for k in x]
})
```

```python
import altair as alt
df = pd.concat([b1,b2])
df = df.melt(id_vars=["x", "params"], value_vars=["PMF", "CDF"], var_name="type", value_name="y")
alt.Chart(df).mark_bar(
    opacity=1,
    size=10,
    fill="#3498db"
).encode(
    x=alt.X("x", title="k"),
    y="y"
).properties(
    width=300,
    height=150
).facet(
    row=alt.Row("params", title=None),
    column=alt.Column("type", sort=["PMF", "CDF"], title="Binomialverteilung")
).configure_header(
    labelFontSize=12,
    titleFontSize=16
).configure_axis(
    grid=False
)
```
