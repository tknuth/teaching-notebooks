```python
from anagram_checker import *
from timeit import timeit
import string
import altair as alt
import pandas as pd
import random
import itertools as it

funcs = (
    tick_while,
    tick_for,
    sort,
    counter,
    collections_counter,
    # brute_force,
)

def take(n):
    i = 0
    l = []
    letters = it.cycle(string.ascii_lowercase)
    while i < n:
        i += 1
        l.append(next(letters))
    return l

def make_anagram(n):
    a = take(n)
    b = a[:]
    random.shuffle(b)
    return "".join(a), "".join(b)

max_length = 50
step = 1

d = {n: make_anagram(n) for n in range(1, max_length, step)}

df = pd.DataFrame({
    f.__name__: [timeit(lambda: f(*d[n]), number=1000) for n in range(1, max_length, step)] 
    for f in funcs
})

df = df.reset_index().melt(id_vars="index", value_vars=map(lambda f: f.__name__, funcs))
df = df.rename(columns={"variable": "Function", "value": "Time", "index": "Word Length"})

alt.data_transformers.enable("default", max_rows=None)
alt.Chart(df).mark_line().encode(
    x="Word Length",
    y="Time",
    color="Function"
)
```
