```python
from bokeh.plotting import figure, show
from bokeh.models import Range1d
from bokeh.io import output_notebook
import numpy as np

bg_color = "#fafafa"
output_notebook()
```

```python
def scatter_plot(x, y, **kwargs):
  p = figure(background_fill_color=bg_color, **kwargs)
  p.scatter(x, y)
  return p
```

```python
n = 50
noise = 2
intercept = 15
slope = 0.4

np.random.seed(42)
e = np.random.normal(0, noise, size=n)
x_max = 50
x = np.linspace(0, x_max, n)
y = intercept + slope * x + e

p = scatter_plot(x, y)
show(p)
```

```python
b = np.sum((x - np.mean(x)) * (y - np.mean(y))) / np.sum((x - np.mean(x))**2)
a = np.mean(y) - b * np.mean(x)

p = scatter_plot(x, y)
p.line(x, a + b * x, color="red")
show(p)
```

```python
X = np.vstack([np.ones(len(x)), x]).T
b, a = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

p = scatter_plot(x, y)
p.line(x, b + a * x, color="red")
show(p)
```

```python
import math

w = np.zeros(2,)
X = np.vstack([np.ones(len(x)), x])
eta = 0.001
max_iteration = 10000

a_hist = [w[1]]
b_hist = [w[0]]

for i in range(max_iteration):
  error = X.T @ w - y
  gradient = X @ error / n
  w = w - eta * gradient
  a_hist.append(w[1])
  b_hist.append(w[0])
```

```python
p = figure()
p.line(a_hist, b_hist)
p.xaxis.axis_label = "y intercept"
p.yaxis.axis_label = "slope"
show(p)
```

```python
p = scatter_plot(x, y)
p.line(x, w[0] + w[1] * x, color="red")
show(p)
```

```python
from sklearn import linear_model

reg = linear_model.LinearRegression()
reg.fit(x.reshape(-1, 1), y)

a = reg.coef_[0]
b = reg.intercept_

p = scatter_plot(x, y)
p.line(x, b + a * x, color="red")
show(p)
```
