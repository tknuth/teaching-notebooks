```
from scipy.special import comb
import matplotlib.pyplot as plt
import numpy as np

# plt.rcParams["figure.figsize"] = (15,5)

def p(n, k, r, m):
  if any([m > r, m > k, k > n, r > n, m > n]):
    return 0
  return (comb(k,m)*comb(n-k,r-m))/comb(n,r)

def plot(k, r, m):
  x = np.arange(1,100)
  y = [p(n, k, r, m) for n in x]

  print("Maximum Likelihood is at n =", x[np.argmax(y)])
  plt.plot(x, y) #, "bo")
```

```
#@title Variablen { run: "auto" }

# Größe der 1. Stichprobe (alle markiert)
k = 10 #@param { type: "slider", min: 0, max: 20, step: 1 }

# Größe der 2. Stichprobe
r = 20 #@param { type: "slider", min: 0, max: 20, step: 1 }

# Anzahl wiedergefundener, markierter Tiere in der 2. Stichprobe
m = 4 #@param { type: "slider", min: 0, max: 20, step: 1 }

plot(k, r, m)
```
