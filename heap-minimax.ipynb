```python
import math

r = [8, 7, 3, 9, 9, 8, 2, 4, 1, 8, 8, 9, 9, 9, 3, 4]
t = [None] * (len(r) - 1) + r # full binary tree

def minimax(i=0, turn="max"): # max starts
    if i >= len(t) // 2 :
        return t[i]
    children = [i * 2 + 1, i * 2 + 2]
    if turn == "max":
        v = -math.inf
        for j in children:
            v = max(v, minimax(j, turn="min"))
    if turn == "min":
        v = math.inf
        for j in children:
            v = min(v, minimax(j, turn="max"))
    t[i] = v
    return v

minimax(turn="max")
```

```python
from graphviz import Digraph

dot = Digraph()

def get_name(i):
    return f"i={i}, v={t[i]}"

for i, v in enumerate(t):
    name = get_name(i)
    dot.node(name)
    if i < len(t) // 2:
        dot.edge(name, get_name(i * 2 + 1))
        dot.edge(name, get_name(i * 2 + 2))
        
dot
```
